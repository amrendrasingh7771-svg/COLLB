```python
obj = LabelEncoder()
```

```python
df.drop('Customer_ID',axis=1)
```

```python
for col in df.columns:
  if (df[col].dtypes == "object64"):
    df[col] = obj.fit_transform(df[col])
```

```python
le=LabelEncoder()
for col in df.columns:
    if (df[col].dtype == 'object')  &  (col !="Churn") :
        df[col]=le.fit_transform(df[col])
```

```python
df['Annual_Income'] = df['Annual_Income'].replace('not_available', np.nan)
df['Annual_Income'] = df['Annual_Income'].fillna(df['Annual_Income'].mode()[0])
df['Annual_Income'] = df['Annual_Income'].astype(float).astype(int)
```

```python
df['Loan_Tenure'] =df['Loan_Tenure'].astype(float).astype(int)
```

```python
df['Loan_Tenure'] = df['Loan_Tenure'].replace('ten',10)
```

```python
# logistic modle

from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
modle_1 = LogisticRegression()

x = df.drop('Default',axis= 1)
```

```python
x = df.drop('Default',axis= 1)

y = df['Default']
```

```python
x_train ,x_test ,y_train, y_test = train_test_split(x,y,train_size=0.80,random_state=0)
```

```python
modle_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_1 = modle_1.predict(x_test)
```

```python
acc_1=accuracy_score(y_pred_1,y_test)*100
```

```python
modle_2 = DecisionTreeClassifier()
```

```python
modle_2.fit(x_train,y_train)
```

```python
y_pred_2 =modle_2.predict(x_test)
```

```python
acc_2=accuracy_score(y_pred_2,y_test)*100
```

```python
modle_3 = RandomForestClassifier()
modle_3.fit(x_train,y_train)
y_pred_3 = modle_3.predict(x_test)
acc_3=accuracy_score(y_pred_3,y_test)*100
```

```python
print('The accuracy score of Logistic refression is :',acc_1)
print('The accuracy score of Decision Tree classifier is :',acc_2)
print('The accuracy score od Random forest tree is :', acc_3)
```
